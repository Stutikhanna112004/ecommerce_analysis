```python
import matplotlib.pyplot as plt
```

```python
categories = ["A","B","C","D"]
values = [ 24, 46 , 13, 38 ]
max_value = max(values)
colors = ["#77DD77" if value != max_value else "#FFD1DC" for value in values]

plt.bar(categories, values , color = colors, width= 0.8, bottom=0)
plt.title("Bar Chart : Compare Categoies", fontdict={
    "fontsize":16,
    "fontweight" : "bold",
    "color":"red",
    "family" : "Arial"})
plt.xlabel("Category", fontdict={
    "fontsize":16,
    "fontweight" : "bold",
    "color":"red",
    "family" : "Arial"})
plt.ylabel("Value", fontdict={
    "fontsize":16,
    "fontweight" : "bold",
    "color":"red",
    "family" : "Arial"})
plt.show()
    
```

```python
categories = ["English","Hindi","Science","Maths"]
values = [ 48, 32 , 41, 18 ]
max_value = max(values)
min_value = min(values)
colors = [
    "#e52165" if value == max_value else "#a2d5c6" if value == min_value else "#0d1137"
    for value in values
]
plt.bar(categories, values , color = colors, width= 0.8, bottom=0)
plt.title("Bar Chart : Marks out of 50", fontdict={
    "fontsize":16,
    "fontweight" : "bold",
    "color":"red",
    "family" : "Arial"})
plt.xlabel("Subject", fontdict={
    "fontsize":16,
    "fontweight" : "bold",
    "color":"red",
    "family" : "Arial"})
plt.ylabel("Marks", fontdict={
    "fontsize":16,
    "fontweight" : "bold",
    "color":"red",
    "family" : "Arial"})
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

days = pd.date_range("2026-01-01", periods=30)
values = np.cumsum(np.random.normal(0, 1, 30)) + 50

plt.plot(days, values, marker="o", color="darkorange")
plt.title("Line Chart: Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Values")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
labels= ["Product A", "Product B" , "Product C" , "Product D"]
sizes = [38, 25,20,20]
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]
explode = (0.05, 0, 0, 0)
plt.pie(sizes,labels=labels,autopct="%1.1f%%", startangle=90, colors=colors,
    explode=explode,
    shadow=True,
    wedgeprops={"edgecolor": "white", "linewidth": 1}
)
plt.title("Pie Chart: Percentage Share")
plt.show()
```
