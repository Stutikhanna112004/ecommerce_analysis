```python
#Step 6: Duplicate Records
df.duplicated().sum()
```

```python
#unique channel names
df["channelTitle"].unique()
```

```python
df["publishedAt"] = pd.to_datetime(df["publishedAt"])
```

```python
df["Year"]=df["publishedAt"].dt.year

df["Month"]=df["publishedAt"].dt.month_name()

df["Day"]=df["publishedAt"].dt.day_name()

df["Hour"] = df["publishedAt"].dt.hour
```

```python
#Count tags per video
df["tag_count"] = df["tags"].str.count(",") + 1
```

```python
#Which videos received the highest views?
highest_views = df.nlargest(10, "viewCount")[["title", "viewCount"]]
highest_views
```

```python
# Which videos received the highest likes?
highest_likes = df.nlargest(10, "likeCount")[["title", "likeCount"]]
highest_likes
```

```python
#Which videos received the highest comments?
highest_comments = df.nlargest(10, "commentCount")[["title", "commentCount"]]
highest_comments
```

```python
#Which year had the maximum uploads?

yearly_uploads = df.groupby("Year")["video_id"].count()
```

```python
#Which month had the maximum uploads?
monthly_uploads = df.groupby("Month")["video_id"].count()
```

```python
#How many HD videos?
df["definition"].value_counts()
```

```python
#How many videos have captions?
df["caption"].value_counts()
```

```python
#What is the average engagement?
df["Engagement Rate"] = ((df["likeCount"] + df["commentCount"]) / df["viewCount"]) * 100

df["Engagement Rate"].describe()
```

```python
#Which videos have the highest engagement?
df.nlargest(10, "Engagement Rate")[["title", "Engagement Rate"]]
```

```python
#Average views every year
df.groupby("Year")["viewCount"].mean()
```

```python
#Average likes every year
df.groupby("Year")["likeCount"].mean()
```

```python
#Do HD videos get better engagement?
df.groupby("definition")[["viewCount","likeCount","commentCount"]].mean()
```

```python
#Does adding captions improve engagement?
df.groupby("caption")[["viewCount","likeCount","commentCount"]].mean()
```

```python
plt.figure(figsize=(12, 5))
monthly_uploads = monthly_uploads.plot(kind="line", marker="o", color="#2E86AB", linewidth=2)

plt.title("Monthly Uploads Trend", fontsize=14, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Uploads")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("monthly_uploads_trend.png", dpi=150)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
yearly_uploads = yearly_uploads.plot(kind="line", marker="o", color="#2E86AB", linewidth=2)

plt.title("Yearly Uploads Trend", fontsize=14, fontweight="bold")
plt.xlabel("Year")
plt.ylabel("Uploads")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("yearly_uploads_trend.png", dpi=150)
plt.show()
```
