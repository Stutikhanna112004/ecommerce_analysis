```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#to check list of functionality of seaborn
print(dir(sns))
```

```python

ft = sns.load_dataset('flights')
```

```python
ft
```

```python
sns.barplot(data = ft, x="month", y = "passengers", hue="month", palette= "viridis")
plt.xticks(rotation=45)
plt.title("Passengers Distribution")
plt.show
```

```python
sns.boxplot(data = ft, x="month", y = "passengers", hue="month", palette= "viridis")
plt.xticks(rotation=45)
plt.title("Passengers Distribution")
plt.show
```

```python
sns.violinplot(data=ft, x="month", y="passengers",  hue="month", palette= "viridis")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(ft["passengers"],fill=True)
plt.title("Density Plot")
plt.show()
```

```python
sns.histplot(ft["passengers"],fill=True)
plt.title("Density Plot")
plt.show()
```

```python
sns.kdeplot(ft["passengers"],fill=True)
plt.title("Density Plot")
plt.show()
```

```python
pivot = ft.pivot(index="month", columns="year", values="passengers")
sns.heatmap(pivot, cmap="YlGnBu")
plt.title("Flights Heatmap")
plt.show()
```

```python
sns.pairplot(ft)
plt.show
```

```python
sns.countplot(data=ft, x="month",  hue="month", palette= "viridis")
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize=(12,8))

#line plot
plt.subplot(2,2,1)
sns.lineplot(data= ft, x= "year", y="passengers")
plt.show()
```

```python
sns.lineplot(data= ft, x= "year", y="passengers",  hue="month", palette= "viridis")
plt.title("Monthly Passenger Trends")
plt.show()
```
